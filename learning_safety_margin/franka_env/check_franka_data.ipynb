```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
```

```python
data_dir = "data/csv_files_postprocessing/"
oa_dir = data_dir + 'obstacle_avoidance/'
vel_dir = data_dir + 'velocity_limits/'
```

```python
# Obstacle Avoidance

nSafe = 25
nUnsafe = 38
nDaring = 12
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
safe_traj = []
for i in range(0, nSafe):
    fname = oa_dir + 'safe_demonstrations/' + str(i+1) + '_eePosition.txt'
    pos = np.loadtxt(fname, delimiter=',').T
    safe_traj.append(pos)
    ax.plot(pos[:,0], pos[:,1], pos[:,2], 'g')

unsafe_traj = []
for i in range(0, nUnsafe):
    fname = oa_dir + 'unsafe_demonstrations/' + str(i+1) + '_eePosition.txt'
    pos = np.loadtxt(fname, delimiter=',').T
    unsafe_traj.append(pos)
    ax.plot(pos[:,0], pos[:,1], pos[:,2], 'r')

daring_traj = []
for i in range(0, nDaring):
    fname = oa_dir + 'daring_demonstrations/' + str(i+1) + '_eePosition.txt'
    pos = np.loadtxt(fname, delimiter=',').T
    daring_traj.append(pos)
    ax.plot(pos[:,0], pos[:,1], pos[:,2], 'b')


ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('Franka CBF Obstacle Avoidance: Position Data')
plt.show()


```

```python
print(len(safe_traj), len(unsafe_traj), len(daring_traj))
```

```python
fig, ax = plt.subplots(3, 1, sharex=True)
for i in range(0, nSafe):
    pos = safe_traj[i]
    ax[0].plot(pos[:,0],'g')
    ax[1].plot(pos[:,1],'g')
    ax[2].plot(pos[:,2],'g')

for i in range(0, nUnsafe):
    pos = unsafe_traj[i]
    ax[0].plot(pos[:,0],'r')
    ax[1].plot(pos[:,1],'r')
    ax[2].plot(pos[:,2],'r')

for i in range(0, nDaring):
    pos = daring_traj[i]
    ax[0].plot(pos[:,0],'b')
    ax[1].plot(pos[:,1],'b')
    ax[2].plot(pos[:,2],'b')
ax[0].set_ylabel('$x$')
ax[1].set_ylabel('$y$')
ax[2].set_ylabel('$z$')
ax[2].set_xlabel('time')

plt.suptitle('Franka CBF Obstacle Avoidance: Position Data')
```

```python
# Velocity Limits

nSafe = 10
nUnsafe = 10
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
safe_traj = []
for i in range(0, nSafe):
    fname = vel_dir + 'slow/' + str(i+1) + '_eePosition.txt'
    pos = np.loadtxt(fname, delimiter=',').T
    safe_traj.append(pos)
    ax.plot(pos[:,0], pos[:,1], pos[:,2], 'g')

unsafe_traj = []
for i in range(0, nUnsafe):
    fname = vel_dir + 'fast/' + str(i+1) + '_eePosition.txt'
    pos = np.loadtxt(fname, delimiter=',').T
    unsafe_traj.append(pos)
    ax.plot(pos[:,0], pos[:,1], pos[:,2], 'r')

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_title('Franka CBF Velocity Limits: Position Data')


```

```python
fig, ax = plt.subplots(3, 1, sharex=True)
for i in range(0, nSafe):
    pos = safe_traj[i]
    ax[0].plot(pos[:,0],'g')
    ax[1].plot(pos[:,1],'g')
    ax[2].plot(pos[:,2],'g')

for i in range(0, nUnsafe):
    pos = unsafe_traj[i]
    ax[0].plot(pos[:,0],'r')
    ax[1].plot(pos[:,1],'r')
    ax[2].plot(pos[:,2],'r')
ax[0].set_ylabel('$x$')
ax[1].set_ylabel('$y$')
ax[2].set_ylabel('$z$')
ax[2].set_xlabel('time')

plt.suptitle('Franka CBF Velocity Limits: Position Data')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
safe_vel = []
for i in range(0, nSafe):
    fname = vel_dir + 'slow/' + str(i+1) + '_eeVelocity.txt'
    vel = np.loadtxt(fname, delimiter=',').T
    safe_vel.append(vel)
    ax.plot(vel[:,0], vel[:,1], vel[:,2], 'g')

unsafe_vel = []
for i in range(0, nUnsafe):
    fname = vel_dir + 'fast/' + str(i+1) + '_eeVelocity.txt'
    vel = np.loadtxt(fname, delimiter=',').T
    unsafe_vel.append(vel)
    ax.plot(vel[:,0], vel[:,1], vel[:,2], 'r')
ax.set_xlabel('$\dot{x}$')
ax.set_ylabel('$\dot{y}$')
ax.set_zlabel('$\dot{z}$')
ax.set_title('Franka CBF Velocity Limits: Velocity Data')
```
